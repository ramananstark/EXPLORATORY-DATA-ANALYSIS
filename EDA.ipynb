```python
import pandas as pd
import numpy as np
import os
```

```python
path="Sales_Data"
files=[file for file in os.listdir(path)]

all_data=pd.DataFrame()
for file in files:
    data = pd.read_csv("Sales_Data/"+file)
    all_data=pd.concat([all_data,data])

all_data.to_csv("all_data.csv",index=False)
```

```python
all_data = all_data.dropna()
all_data
```

```python
#all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
#print(all_data['Order Date'].dtype)
all_data=all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
all_data['City']= all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data
all_data['state']=all_data['Purchase Address'].apply(lambda x: x.split(',')[2].split(' ')[1])
all_data['City']=all_data['City'] + "-" + all_data['state']
all_data=all_data.drop(columns='state')
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data
```

```python
Products=all_data.groupby('Product').sum().reset_index()
Products
plt.bar(Products['Product'],Products['Quantity Ordered'])
plt.xticks(Products['Product'],rotation='vertical')
plt.xlabel("product")
plt.ylabel("sold_count")
plt.title("Number of products sold by each product")
plt.show()
```

```python
prices=all_data.groupby('Product').mean().reset_index()
product=prices['Product']
price=prices['Price Each']
plt.plot(product,price)
plt.xticks(product,rotation='vertical')
plt.xlabel('products')
plt.ylabel('price')
plt.title('Price of each product')
plt.show()
```
